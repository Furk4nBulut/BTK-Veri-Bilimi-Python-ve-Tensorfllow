# visualization (high level)

```python
nparr = np.linspace(0,10,20)
nparr2=nparr**2
```

```python
(myFigure,myAxes)=plt.subplots()
```

```python
(myFigure,myAxes)=plt.subplots()
myAxes.plot(nparr,nparr+2,"blue",linewidth=5.0)
myAxes.plot(nparr,nparr+3,color="pink",linewidth=6.0,linestyle=":")

myAxes.plot(nparr,nparr+4,color="yellow",linestyle="--")
myAxes.plot(nparr,nparr+5,color="red",linestyle="-.")

myAxes.plot(nparr,nparr+8,color="orange",linestyle="--"
            ,marker="o",markerfacecolor="b",markersize=6)
myAxes.plot(nparr,nparr+9,color="orange",linestyle="--"
            ,marker="+",markerfacecolor="b",markersize=15)
```

```python
(myFigure,myAxes)=plt.subplots()
myAxes.plot(nparr,nparr2,"#c96f23",alpha=0.5)
myAxes.plot(nparr2,nparr,color="#3a95a8",alpha=0.3)
```
